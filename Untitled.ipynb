```python
from __future__ import annotations
from pathlib import Path
from typing import Dict, Any, List, Optional
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Giúp biểu đồ hiển thị ngay trong Notebook
%matplotlib inline 
sns.set_style("whitegrid")
```

```python
def _ensure_output_dir(output_dir: str | Path) -> Path:
    path = Path(output_dir)
    path.mkdir(parents=True, exist_ok=True)
    return path


def _validate_columns(df: pd.DataFrame, columns: List[str]) -> None:
    missing = [c for c in columns if c not in df.columns]
    if missing:
        raise ValueError(f"Missing columns: {missing}")


def summarize_numeric_feature(df: pd.DataFrame, column: str) -> Dict[str, Any]:
    """Return descriptive statistics and quality checks for one numeric column."""
    if column not in df.columns:
        raise ValueError(f"Column '{column}' does not exist")
    if not pd.api.types.is_numeric_dtype(df[column]):
        raise TypeError(f"Column '{column}' is not numeric")

    s = df[column].dropna()
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outlier_mask = (s < lower) | (s > upper)

    return {
        "feature": column,
        "count": int(s.shape[0]),
        "missing_count": int(df[column].isna().sum()),
        "missing_pct": float(df[column].isna().mean() * 100),
        "mean": float(s.mean()),
        "median": float(s.median()),
        "std": float(s.std()),
        "min": float(s.min()),
        "q1": float(q1),
        "q3": float(q3),
        "max": float(s.max()),
        "p95": float(s.quantile(0.95)),
        "p99": float(s.quantile(0.99)),
        "skewness": float(s.skew()),
        "kurtosis": float(s.kurt()),
        "zero_count": int((s == 0).sum()),
        "negative_count": int((s < 0).sum()),
        "iqr": float(iqr),
        "outlier_lower": float(lower),
        "outlier_upper": float(upper),
        "outlier_count": int(outlier_mask.sum()),
        "outlier_pct": float(outlier_mask.mean() * 100),
    }


def build_numeric_summary_table(
    df: pd.DataFrame,
    columns: List[str],
    output_dir: Optional[str | Path] = None,
) -> pd.DataFrame:
    """Build summary table for multiple numeric columns."""
    _validate_columns(df, columns)
    summary = pd.DataFrame([summarize_numeric_feature(df, c) for c in columns])

    if output_dir is not None:
        out_dir = _ensure_output_dir(output_dir)
        summary.to_csv(out_dir / "numeric_summary.csv", index=False)

    return summary


def compare_numeric_by_target(
    df: pd.DataFrame,
    numeric_columns: List[str],
    target_column: str = "is_canceled",
    output_dir: Optional[str | Path] = None,
) -> Dict[str, pd.DataFrame]:
    """Create grouped descriptive statistics by target for numeric columns."""
    _validate_columns(df, numeric_columns + [target_column])

    outputs: Dict[str, pd.DataFrame] = {}
    for col in numeric_columns:
        grouped = (
            df.groupby(target_column)[col]
            .agg(["count", "mean", "median", "std", "min", "max"])
            .reset_index()
        )
        outputs[col] = grouped

        if output_dir is not None:
            out_dir = _ensure_output_dir(output_dir)
            grouped.to_csv(out_dir / f"{col}_by_{target_column}.csv", index=False)

    return outputs


def detect_outliers_iqr(
    df: pd.DataFrame,
    column: str,
    target_column: Optional[str] = None,
) -> Dict[str, Any]:
    """Detect outliers using IQR rule and optionally summarize by target."""
    _validate_columns(df, [column])
    s = df[column].dropna()
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outlier_mask_full = (df[column] < lower) | (df[column] > upper)
    result: Dict[str, Any] = {
        "feature": column,
        "lower_bound": float(lower),
        "upper_bound": float(upper),
        "outlier_count": int(outlier_mask_full.sum()),
        "outlier_pct": float(outlier_mask_full.mean() * 100),
        "sample_outliers": df.loc[outlier_mask_full, [column]].head(10).to_dict(orient="records"),
    }

    if target_column is not None:
        _validate_columns(df, [target_column])
        by_target = (
            df.assign(is_outlier=outlier_mask_full)
            .groupby(target_column)["is_outlier"]
            .mean()
            .mul(100)
            .reset_index(name="outlier_pct")
        )
        result["outlier_pct_by_target"] = by_target.to_dict(orient="records")

    return result


def plot_numeric_distribution(
    df: pd.DataFrame,
    column: str,
    output_path: str | Path,
    bins: int = 40,
) -> str:
    """Save histogram and boxplot for one numeric feature."""
    _validate_columns(df, [column])
    output_path = str(output_path)

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    axes[0].hist(df[column].dropna(), bins=bins)
    axes[0].set_title(f"Distribution of {column}")
    axes[0].set_xlabel(column)
    axes[0].set_ylabel("Frequency")

    sns.boxplot(x=df[column].dropna(), ax=axes[1])
    axes[1].set_title(f"Boxplot of {column}")
    axes[1].set_xlabel(column)

    plt.tight_layout()
    plt.savefig(output_path, dpi=200, bbox_inches="tight")
    plt.close(fig)
    return output_path


def plot_numeric_by_target(
    df: pd.DataFrame,
    column: str,
    target_column: str,
    output_path: str | Path,
) -> str:
    """Save boxplot comparison of one numeric feature by target class."""
    _validate_columns(df, [column, target_column])
    output_path = str(output_path)

    fig, ax = plt.subplots(figsize=(8, 5))
    sns.boxplot(data=df, x=target_column, y=column, ax=ax)
    ax.set_title(f"{column} by {target_column}")
    ax.set_xlabel(target_column)
    ax.set_ylabel(column)

    plt.tight_layout()
    plt.savefig(output_path, dpi=200, bbox_inches="tight")
    plt.close(fig)
    return output_path


def run_member4_numeric_eda(
    df: pd.DataFrame,
    output_dir: str | Path,
    numeric_columns: Optional[List[str]] = None,
    target_column: str = "is_canceled",
) -> Dict[str, Any]:
    """Run the full Member 4 pipeline for numeric EDA."""
    if numeric_columns is None:
        numeric_columns = ["lead_time", "adr"]

    _validate_columns(df, numeric_columns + [target_column])
    out_dir = _ensure_output_dir(output_dir)

    summary_table = build_numeric_summary_table(df, numeric_columns, output_dir=out_dir)
    grouped = compare_numeric_by_target(df, numeric_columns, target_column, output_dir=out_dir)

    outlier_results = {}
    plot_paths = {}
    for col in numeric_columns:
        outlier_results[col] = detect_outliers_iqr(df, col, target_column=target_column)
        plot_paths[f"{col}_distribution"] = plot_numeric_distribution(
            df, col, out_dir / f"{col}_distribution.png"
        )
        plot_paths[f"{col}_by_target"] = plot_numeric_by_target(
            df, col, target_column, out_dir / f"{col}_by_target.png"
        )

    return {
        "summary_table": summary_table.to_dict(orient="records"),
        "grouped_stats": {k: v.to_dict(orient="records") for k, v in grouped.items()},
        "outliers": outlier_results,
        "plots": plot_paths,
    }
```
